# Dataset de Churn de Biblioteca

## Importando bibliotecas

```python
import numpy as np
import pandas as pd
```

## Criando Dataframe simulado

```python
tipo_usuario = ['Graduação', 'Pós-graduação','Docência','Colaborador']

institutos = ['Artes','Biologia','Computação','Economia','Estudos da Linguagem','Filosofia e Ciências Humanas','Física',
              'Geociências','Matemática, Estatística e Computação Científica','Química']

tamanho_dados_uns = 95
tamanho_dados_zeros = 5

data = {
    'nivel_matricula': np.random.choice(tipo_usuario, size=tamanho_dados_uns),
    'institutos': np.random.choice(institutos, size=tamanho_dados_uns),
    
    'mediana_emprestimos': np.round(np.random.normal(loc=25, scale=10, size=tamanho_dados_uns)).astype(int),
    'mediana_acessos_biblioteca_digital': np.round(np.random.normal(loc=30, scale=15, size=tamanho_dados_uns)).astype(int),
    'mediana_participacoes_eventos': np.random.randint(0, 5, size=tamanho_dados_uns),
    'mediana_consultas': np.round(np.random.normal(loc=15, scale=7.5, size=tamanho_dados_uns)).astype(int),
    'mediana_visitas': np.round(np.random.normal(loc=15, scale=7.5, size=tamanho_dados_uns)).astype(int),
    'duracao_filiacao': np.random.randint(0, 1825, size=tamanho_dados_uns),
    'satisfacao': np.random.randint(0, 6, size=tamanho_dados_uns),
    'mediana_reclamacoes': np.random.randint(0, 4, size=tamanho_dados_uns),
    'mediana_sugestoes': np.random.randint(0, 4, size=tamanho_dados_uns)
}

data_zeros = {
    'nivel_matricula': np.random.choice(tipo_usuario, size=tamanho_dados_zeros),
    'institutos': np.random.choice(institutos, size=tamanho_dados_zeros),
    
    'mediana_emprestimos': np.zeros(tamanho_dados_zeros, dtype=int),
    'mediana_acessos_biblioteca_digital': np.zeros(tamanho_dados_zeros, dtype=int),
    'mediana_participacoes_eventos': np.zeros(tamanho_dados_zeros, dtype=int),
    'mediana_consultas': np.zeros(tamanho_dados_zeros, dtype=int),
    'mediana_visitas': np.zeros(tamanho_dados_zeros, dtype=int),
    'duracao_filiacao': np.random.randint(0, 1825, size=tamanho_dados_zeros),
    'satisfacao': np.random.randint(0, 6, size=tamanho_dados_zeros),
    'mediana_reclamacoes': np.random.randint(0, 4, size=tamanho_dados_zeros),
    'mediana_sugestoes': np.random.randint(0, 4, size=tamanho_dados_zeros)
}

df_zeros = pd.DataFrame(data_zeros)

# Cria o dataframe
df_uns = pd.DataFrame(data)

df = pd.concat([df_zeros,df_uns])

# Marca a coluna de churn
df['churn'] = np.where((df['mediana_emprestimos']==0) & 
                       (df['mediana_acessos_biblioteca_digital']==0) & 
                       (df['mediana_participacoes_eventos']==0) & 
                       (df['mediana_consultas']==0) & 
                       (df['mediana_visitas']==0), 'sim', 'não')

df
```
