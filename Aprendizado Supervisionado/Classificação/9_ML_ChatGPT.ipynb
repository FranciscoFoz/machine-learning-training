```python
import pandas as pd
import numpy as np
from scipy.stats import zscore
from sklearn.preprocessing import MinMaxScaler

def preprocess_data(df):
    # Remover as colunas "ID", "ZIP Code" e "Experience"
    df = df.drop(columns=["ID", "ZIP Code", "Experience"])
    
    # Retirar outliers da coluna "Mortgage" usando zscore
    z_scores = zscore(df["Mortgage"])
    df = df[(np.abs(z_scores) < 3)]
    
    # Multiplicar os valores da coluna "CCAvg" por 12
    df["CCAvg"] = df["CCAvg"] * 12
    
    # Normalizar as colunas "Age", "Income", "Family", "CCAvg" e "Mortgage" usando MinMaxScaler
    scaler = MinMaxScaler()
    columns_to_normalize = ["Age", "Income", "Family", "CCAvg", "Mortgage"]
    df[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])
    
    # Aplicar a técnica de one-hot encoding na coluna "Education" usando get_dummies
    df = pd.get_dummies(df, columns=["Education"])
    
    return df

import pandas as pd

url = "https://github.com/alura-cursos/machine_learning_chatgpt/blob/main/Dados/emprestimo.xlsx?raw=true"
df = pd.read_excel(url)

df_processado = preprocess_data(df)
df_processado
```

    Escreva o código Python que seleciona todas as nossas features do DataFrame "df", exceto o target ("Personal Loan"), em uma variável de nome X.


    A variável X contém as features e  a variável y contém o target.
    Quero separar os dados em dados de treino e de teste. 
    Separe 20% dos dados para teste.

```python
from sklearn.model_selection import train_test_split

# Separar as features (X) e o target (y)
X = df_processado.drop("Personal Loan", axis=1)
y = df_processado["Personal Loan"]

# Separar os dados em treinamento e teste, preservando a proporção entre as classes
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Verificar o tamanho dos conjuntos de treino e teste
print("Tamanho do conjunto de treino:", len(X_train))
print("Tamanho do conjunto de teste:", len(X_test))
```

    Escreva o código com a Sklearn para fazer o treinamento do modelo de Machine Learning em um problema de classificação com um modelo baseline, como o Dummy Classifier.

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

# Criar e treinar o Dummy Classifier
dummy_clf = DummyClassifier(strategy="most_frequent")  # Estratégia de previsão mais frequente
dummy_clf.fit(X_train, y_train)

# Fazer previsões no conjunto de teste
y_pred = dummy_clf.predict(X_test)

# Calcular a acurácia do modelo
accuracy = accuracy_score(y_test, y_pred)
print("Acurácia do Dummy Classifier:", accuracy)
```

```python
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler


# Treinar e avaliar o Dummy Classifier
dummy_clf = DummyClassifier(strategy="most_frequent")
dummy_clf.fit(X_train, y_train)
y_pred_dummy = dummy_clf.predict(X_test)
accuracy_dummy = accuracy_score(y_test, y_pred_dummy)
print("Acurácia do Dummy Classifier:", accuracy_dummy)

# Treinar e avaliar a Logistic Regression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred_logreg = logreg.predict(X_test)
accuracy_logreg = accuracy_score(y_test, y_pred_logreg)
print("Acurácia da Logistic Regression:", accuracy_logreg)

# Treinar e avaliar a Decision Tree
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
accuracy_dt = accuracy_score(y_test, y_pred_dt)
print("Acurácia da Decision Tree:", accuracy_dt)
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

# Treinar e avaliar o Random Forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print("Acurácia do Random Forest:", accuracy_rf)

# Treinar e avaliar o Gradient Boosting
gb = GradientBoostingClassifier()
gb.fit(X_train, y_train)
y_pred_gb = gb.predict(X_test)
accuracy_gb = accuracy_score(y_test, y_pred_gb)
print("Acurácia do Gradient Boosting:", accuracy_gb)

# Treinar e avaliar as Redes Neurais Artificiais
mlp = MLPClassifier()
mlp.fit(X_train, y_train)
y_pred_mlp = mlp.predict(X_test)
accuracy_mlp = accuracy_score(y_test, y_pred_mlp)
print("Acurácia das Redes Neurais Artificiais:", accuracy_mlp)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Função para plotar a matriz de confusão
def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred, normalize='true')
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='.2', cmap='Blues')
    plt.title(title)
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()

# Obtendo as previsões dos modelos
dummy_preds = dummy_clf.predict(X_test)
logreg_preds = logreg.predict(X_test)
dt_preds = dt.predict(X_test)

rf_preds = rf.predict(X_test)
gb_preds = gb.predict(X_test)

# Plotando as matrizes de confusão
plot_confusion_matrix(y_test, dummy_preds, title='Confusion Matrix - Dummy Classifier')
plot_confusion_matrix(y_test, logreg_preds, title='Confusion Matrix - Logistic Regression')
plot_confusion_matrix(y_test, dt_preds, title='Confusion Matrix - Decision Tree')
plot_confusion_matrix(y_test, rf_preds, title='Confusion Matrix - Random Forest')
plot_confusion_matrix(y_test, gb_preds, title='Confusion Matrix - Gradient Boosting')
```
