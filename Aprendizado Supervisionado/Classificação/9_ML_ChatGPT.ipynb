```python
import pandas as pd
import numpy as np
from scipy.stats import zscore
from sklearn.preprocessing import MinMaxScaler

def preprocess_data(df):
    # Remover as colunas "ID", "ZIP Code" e "Experience"
    df = df.drop(columns=["ID", "ZIP Code", "Experience"])
    
    # Retirar outliers da coluna "Mortgage" usando zscore
    z_scores = zscore(df["Mortgage"])
    df = df[(np.abs(z_scores) < 3)]
    
    # Multiplicar os valores da coluna "CCAvg" por 12
    df["CCAvg"] = df["CCAvg"] * 12
    
    # Normalizar as colunas "Age", "Income", "Family", "CCAvg" e "Mortgage" usando MinMaxScaler
    scaler = MinMaxScaler()
    columns_to_normalize = ["Age", "Income", "Family", "CCAvg", "Mortgage"]
    df[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])
    
    # Aplicar a técnica de one-hot encoding na coluna "Education" usando get_dummies
    df = pd.get_dummies(df, columns=["Education"])
    
    return df

import pandas as pd

url = "https://github.com/alura-cursos/machine_learning_chatgpt/blob/main/Dados/emprestimo.xlsx?raw=true"
df = pd.read_excel(url)

df_processado = preprocess_data(df)
df_processado
```

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

# Criar e treinar o Dummy Classifier
dummy_clf = DummyClassifier(strategy="most_frequent")  # Estratégia de previsão mais frequente
dummy_clf.fit(X_train, y_train)

# Fazer previsões no conjunto de teste
y_pred = dummy_clf.predict(X_test)

# Calcular a acurácia do modelo
accuracy = accuracy_score(y_test, y_pred)
print("Acurácia do Dummy Classifier:", accuracy)
```

```python
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler


# Treinar e avaliar o Dummy Classifier
dummy_clf = DummyClassifier(strategy="most_frequent")
dummy_clf.fit(X_train, y_train)
y_pred_dummy = dummy_clf.predict(X_test)
accuracy_dummy = accuracy_score(y_test, y_pred_dummy)
print("Acurácia do Dummy Classifier:", accuracy_dummy)

# Treinar e avaliar a Logistic Regression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred_logreg = logreg.predict(X_test)
accuracy_logreg = accuracy_score(y_test, y_pred_logreg)
print("Acurácia da Logistic Regression:", accuracy_logreg)

# Treinar e avaliar a Decision Tree
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
accuracy_dt = accuracy_score(y_test, y_pred_dt)
print("Acurácia da Decision Tree:", accuracy_dt)
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

# Treinar e avaliar o Random Forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print("Acurácia do Random Forest:", accuracy_rf)

# Treinar e avaliar o Gradient Boosting
gb = GradientBoostingClassifier()
gb.fit(X_train, y_train)
y_pred_gb = gb.predict(X_test)
accuracy_gb = accuracy_score(y_test, y_pred_gb)
print("Acurácia do Gradient Boosting:", accuracy_gb)

# Treinar e avaliar as Redes Neurais Artificiais
mlp = MLPClassifier()
mlp.fit(X_train, y_train)
y_pred_mlp = mlp.predict(X_test)
accuracy_mlp = accuracy_score(y_test, y_pred_mlp)
print("Acurácia das Redes Neurais Artificiais:", accuracy_mlp)
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Definir as métricas a serem calculadas
metrics = ["accuracy", "precision", "recall", "F1-score", "ROC AUC"]

# Definir os modelos e suas respectivas previsões e rótulos verdadeiros
models = ["dummy_model", "logreg_model", "dt_model", "random_forest_model", "gradient_boosting_model"]
predictions = [dummy_clf.predict(X_test), logreg.predict(X_test), dt.predict(X_test),
               rf.predict(X_test), gb.predict(X_test)]
true_labels = y_test

# Inicializar um dicionário para armazenar as métricas de cada modelo
metrics_dict = {}

# Calcular as métricas para cada modelo
for model, prediction in zip(models, predictions):
    accuracy = accuracy_score(true_labels, prediction)
    precision = precision_score(true_labels, prediction)
    recall = recall_score(true_labels, prediction)
    f1 = f1_score(true_labels, prediction)
    roc_auc = roc_auc_score(true_labels, prediction)
    
    metrics_dict[model] = [accuracy, precision, recall, f1, roc_auc]

# Criar o DataFrame
metrics_df = pd.DataFrame.from_dict(metrics_dict, orient='index', columns=metrics)
metrics_df = metrics_df.round(2)  # Arredondar as métricas para duas casas decimais

# Exibir o DataFrame
metrics_df
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np

# Função para plotar a curva ROC
def plot_roc_curve(y_true, y_score, label):
    fpr, tpr, _ = roc_curve(y_true, y_score)
    roc_auc = auc(fpr, tpr)

    plt.plot(fpr, tpr, label='{} (AUC = {:.2f})'.format(label, roc_auc))
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('Taxa de Falso Positivo')
    plt.ylabel('Taxa de Verdadeiro Positivo')
    plt.title('Curva ROC')
    plt.legend(loc='lower right')

# Realizar a validação cruzada e plotar a curva ROC para cada modelo
plt.figure(figsize=(12,8))
for model, y_score in [('Decision Tree', dt.predict_proba(X_test)[:, 1]),
                       ('Random Forest', rf.predict_proba(X_test)[:, 1]),
                       ('Gradient Boosting', gb.predict_proba(X_test)[:, 1])]:
    plot_roc_curve(y_test, y_score, model)

# Exibir o gráfico da curva ROC
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.metrics import accuracy_score
from imblearn.over_sampling import SMOTE

# Separar as features (X) e o target (y)
X = df.drop("Personal Loan", axis=1)
y = df["Personal Loan"]

# Aplicar balanceamento de dados com SMOTE
smote = SMOTE()
X_balanced, y_balanced = smote.fit_resample(X, y)

# Dividir os dados em treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X_balanced, y_balanced, test_size=0.2, random_state=42)

# Treinar e avaliar o Decision Tree
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
accuracy_dt = accuracy_score(y_test, y_pred_dt)
print("Acurácia do Decision Tree após balanceamento:", accuracy_dt)

# Treinar e avaliar o Gradient Boosting
gb = GradientBoostingClassifier()
gb.fit(X_train, y_train)
y_pred_gb = gb.predict(X_test)
accuracy_gb = accuracy_score(y_test, y_pred_gb)
print("Acurácia do Gradient Boosting após balanceamento:", accuracy_gb)

# Treinar e avaliar o Random Forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print("Acurácia do Random Forest após balanceamento:", accuracy_rf)
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Função para plotar a matriz de confusão
def plot_confusion_matrix(y_true, y_pred, classes):
    cm = confusion_matrix(y_true, y_pred,normalize='true')
    sns.heatmap(cm, annot=True, fmt='.2', cmap='Blues')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')

# Obter a matriz de confusão para o Decision Tree
plt.figure(figsize=(8, 6))
plot_confusion_matrix(y_test, y_pred_dt, classes=['Não Aceitou', 'Aceitou'])
plt.title('Confusion Matrix - Decision Tree')
plt.show()

# Obter a matriz de confusão para o Random Forest
plt.figure(figsize=(8, 6))
plot_confusion_matrix(y_test, y_pred_rf, classes=['Não Aceitou', 'Aceitou'])
plt.title('Confusion Matrix - Random Forest')
plt.show()

# Obter a matriz de confusão para o Gradient Boosting
plt.figure(figsize=(8, 6))
plot_confusion_matrix(y_test, y_pred_gb, classes=['Não Aceitou', 'Aceitou'])
plt.title('Confusion Matrix - Gradient Boosting')
plt.show()
```

```python
from sklearn.model_selection import RepeatedStratifiedKFold, cross_validate
from sklearn.metrics import roc_auc_score
import numpy as np

# Importar RepeatedStratifiedKFold e cross_validate

from sklearn.model_selection import RepeatedStratifiedKFold, cross_validate

# Crie uma variável para definir a quantidade de divisões e repetições
n_splits = 5
n_repeats = 3

# Criar o objeto RepeatedStratifiedKFold
rskf = RepeatedStratifiedKFold(n_splits=n_splits, n_repeats=n_repeats, random_state=42)

# Lista para armazenar os resultados de cada modelo
results = []

# Realizar a validação cruzada para cada modelo
for model, clf in [('Random Forest', rf), ('Decision Tree', dt), ('Gradient Boosting', gb)]:
    # Realizar a validação cruzada e obter os valores de AUC para cada repetição
    cv_results = cross_validate(clf, X_balanced, y_balanced, cv=rskf, scoring='roc_auc', return_train_score=True)

    # Obter as médias e desvios padrão dos valores de AUC para os dados de treino e teste
    train_auc_mean = np.mean(cv_results['train_score'])
    train_auc_std = np.std(cv_results['train_score'])
    test_auc_mean = np.mean(cv_results['test_score'])
    test_auc_std = np.std(cv_results['test_score'])

    # Adicionar os resultados à lista
    results.append((model, train_auc_mean, train_auc_std, test_auc_mean, test_auc_std))

# Exibir os resultados da validação cruzada
print("Resultado da Validação Cruzada:")
df_results = pd.DataFrame(results,columns=["Modelo", "Train AUC (Média)", "Train AUC (Desvio)", "Test AUC (Média)", "Test AUC (Desvio)"])
df_results
```
