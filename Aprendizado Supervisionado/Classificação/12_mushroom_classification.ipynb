```python
variavel_alvo = 'class'
variaveis = [col for col in df.columns if col != variavel_alvo]

for variavel in variaveis:
    fig = px.histogram(df,x=variavel,
             text_auto=True,
             width=500,
             color='class',
             barmode='group',
             title=f'ComparaÃ§Ã£o de {variavel} vs {variavel_alvo}')
    fig.show()
```
